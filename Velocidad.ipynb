```python
Distancia = float(input("Ingresar la distancia recorrida (en metros): "))
Tiempo = float(input("Ingresar el tiempo empleado (en esgundos): "))
Velocidad = Distancia / Tiempo
print(f"La velocidad es {Velocidad} m/s")
```
